**1.** Define una clase, que se llame Estudiantes_Posgrado_IMP, con los siguientes atributos:

nombre, apellidos, correo, nivel y status

Se requiere que el atributo status tenga un valor por default de "Reprobado"

Nota nivel se refiere a "Maestría" o "Doctorado"

Crear un objeto que se llame estudiante, con tus datos personales. 

```python
class Estudiantes_Posgrado_IMP:
    def __init__(self, nombre, apellidos, correo, nivel, status="Reprobado"):
        self.nombre = nombre
        self.apellidos = apellidos
        self.correo = correo
        self.nivel = nivel
        self.status = status

estudiante = Estudiantes_Posgrado_IMP(nombre="Eduardo Daniel",apellidos="Prado Rodríguez",correo="eduardo.prado@gmail.com",nivel="Doctorado")

print(f"Estudiante: {estudiante.nombre} {estudiante.apellidos}")
print(f"Nivel: {estudiante.nivel}")
print(f"Correo: {estudiante.correo}")
print(f"Status actual: {estudiante.status}")
```

**2.**  Crea un método que se llame resultado, dentro de la clase Estudiantes_Posgrado_IMP, que cambie el atributo status de reprobado a aprobado si la calificacion es mayor o igual a 8. Nota, calificacion seria un parámetro de entrada del método resultado.  Después de definir esta clase, crear un objeto estudiante y aplicar el método resultado considerando una calificación de 10. Imprimir el status final del objeto estudiante.



```python
class Estudiantes_Posgrado_IMP:
    def __init__(self, nombre, apellidos, correo, nivel, status="Reprobado"):
        self.nombre = nombre
        self.apellidos = apellidos
        self.correo = correo
        self.nivel = nivel
        self.status = status
        
    def resultado(self, calificacion):
        if calificacion >= 8:
            self.status = "Aprobado"

estudiante = Estudiantes_Posgrado_IMP(nombre="Eduardo Daniel",apellidos="Prado Rodríguez",correo="eduardo.prado@gmail.com",nivel="Doctorado")
estudiante.resultado(9)
print(f"Status final: {estudiante.status}")
```
