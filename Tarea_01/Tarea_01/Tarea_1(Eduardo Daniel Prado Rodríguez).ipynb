Ejercicio 3. Conversor de Temperaturas

1. Pide al usuario que ingrese una temperatura en °C.
2. Convierte esa temperatura a °F usando la fórmula:  
   $$ F = C \times \frac{9}{5} + 32 $$
3. Muestra el resultado con un mensaje claro redondeando a 2 decimales.


```python
# Respuesta

temperatura_c = float(input("Ingresa una temperatura en °C: "))
temperatura_f = (temperatura_c * 9/5) + 32
print(f"La temperatura de {temperatura_c}°C es igual a {temperatura_f:.2f}°F.")
```

Ejercicio 4. Escribe un programa que pida números al usuario y los vaya sumando. El ciclo termina cuando el usuario ingresa un `0` o cuando la suma es mayor que 50. Usa `while` y `break`.


```python
# Respuesta
suma = 0
while suma <= 50:
    numero = int(input("Ingresa un número (0 para salir): "))
    if numero == 0:
        print("Ingresaste 0. Fin del ciclo")
        break
    suma += numero
    print(f"Suma acumulada actual: {suma}")
    if suma > 50:
        print("Fin del ciclo.")
print(f"La suma final es: {suma}")
```

Ejercicio 5. Usa un ciclo for para guardar en una lista los números del 1 al 10, pero salta los números pares usando continue.



```python
# Respuesta

numeros_impares = []
for i in range(1, 11):
    if i % 2 == 0:
        continue  
    numeros_impares.append(i)

print("Lista resultante:", numeros_impares)
```

Ejercicio 7. Escribe una función *my_max(x)* que de como salida el valor máximo (más grande) de una lista *x*. No use la función de Python *max*.

Encuentra el máximo de la siguiente lista 

x = [10, 56, 258, 16, 24, 18, 265, 893, 52, 39, 82]

```python
# Respuesta
def my_max(x):
    maximo = x[0]
    for i in x:
        if i > maximo:
            maximo = i
    return maximo
    
x = [10, 56, 258, 16, 24, 18, 265, 893, 52, 39, 82]

numero_maximo = my_max(x)
print(f"El número máximo es: ", numero_maximo)
```

Ejercicio 8. Pide al usuario una contraseña. Permite hasta 3 intentos, y termina el ciclo si la acierta. Usa while, break y un contador.

```python
# Respuesta
contrasena_correcta = "Eduardo2594"

intentos = 0

while intentos < 3:
    intento_usuario = input("Ingresa la contraseña: ")
    intentos += 1
    if intento_usuario == contrasena_correcta:
        print("Contraseña correcta")
        break  
    else:
        intentos_restantes = 3 - intentos
        if intentos_restantes > 0:
            print(f"Incorrecto. Te quedan {intentos_restantes} intentos.")
        else:
            print("Has agotado tus 3 intentos. Acceso bloqueado.")
```
