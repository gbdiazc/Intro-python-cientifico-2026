Ejercicio 7. Escribe una función *my_max(x)* que de como salida el valor máximo (más grande) de una lista *x*. No use la función de Python *max*.

Encuentra el máximo de la siguiente lista 

x = [10, 56, 258, 16, 24, 18, 265, 893, 52, 39, 82]

```python
# Respuesta
x = [10, 56, 258, 16, 24, 18, 265, 893, 52, 39, 82]
def my_max(x):
    my_max=x[0]
    for i in x:
        if my_max < i:
            my_max = i
    return my_max
my_max(x)
```

Ejercicio 8. Pide al usuario una contraseña. Permite hasta 3 intentos, y termina el ciclo si la acierta. Usa while, break y un contador.

```python
# Respuesta
contraseña_correcta = "posimp"
intentos_maximos = 3
contador_intentos = 0

while contador_intentos < intentos_maximos:
    intento = input("Introduce la contraseña: ")
    contador_intentos += 1   
    
    if intento == contraseña_correcta:
        print("Acceso concedido.")
        break 
    else:
        intentos_restantes = intentos_maximos - contador_intentos
        if intentos_restantes > 0:
            print(f"Contraseña incorrecta. Te quedan {intentos_restantes} intentos.")
        else:
            print("Has agotado tus 3 intentos. Acceso bloqueado.")
```
