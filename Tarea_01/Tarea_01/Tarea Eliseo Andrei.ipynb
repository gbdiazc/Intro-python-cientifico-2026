Ejercicio 3. Conversor de Temperaturas

1. Pide al usuario que ingrese una temperatura en °C.
2. Convierte esa temperatura a °F usando la fórmula:  
   $$ F = C \times \frac{9}{5} + 32 $$
3. Muestra el resultado con un mensaje claro redondeando a 2 decimales.

---
Solucion

```python
print('Estan es un conversor de temperatura de °C a °F \n')

opci='si'
while opci=='si':    
    temp=float(input('Ingresa tu temperatura en °C: '))
    f=(temp*(9/5))+(32)
    print('Tu temperatura es {}°F'.format(f))
    opci=str(input('¿Quieres hacer otra conversion? si no: '))
    if opci=='no':
        print('Adios, usuario.')
        break
```

Ejercicio 4. Escribe un programa que pida números al usuario y los vaya sumando. El ciclo termina cuando el usuario ingresa un `0` o cuando la suma es mayor que 50. Usa `while` y `break`.

---
solucion

```python
print('Este codigo suma los numeros que vayas ingresando. Este se detiene cuando ingreses el numero 0.')
ing=float(input('Ingresa tu primer numero a sumar: '))
suma=0
while ing!=0:
    suma= suma+ing
    print('Tu suma ya es: {}'.format(suma))
    ing=float(input('Ingresa tu otro numero a sumar: '))
    if ing==0:
        print('Has ingresado el 0, la calculadora se detiene.')
        break
```

Ejercicio 5. Usa un ciclo for para guardar en una lista los números del 1 al 10, pero salta los números pares usando continue.

---
solucion

```python
lista=[]
for i in range (1,11):
    if i%2==0:
        continue
    lista.append(i)
print(lista)
```

Ejercicio 7. Escribe una función *my_max(x)* que de como salida el valor máximo (más grande) de una lista *x*. No use la función de Python *max*.

Encuentra el máximo de la siguiente lista

x = [10, 56, 258, 16, 24, 18, 265, 893, 52, 39, 82]

---
solucion

```python
def my_max(x):
    maximo=0
    for i in (x):
        if i > maximo:
            maximo = i

    return maximo
```

```python
x = [10, 56, 258, 16, 24, 18, 265, 893, 52, 39, 82]
my_max(x)
```
