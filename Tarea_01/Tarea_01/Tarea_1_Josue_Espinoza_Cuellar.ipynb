Ejercicio 3. Conversor de Temperaturas

1. Pide al usuario que ingrese una temperatura en °C.
2. Convierte esa temperatura a °F usando la fórmula:  
   $$ F = C \times \frac{9}{5} + 32 $$
3. Muestra el resultado con un mensaje claro redondeando a 2 decimales.


```python
# Respuesta
celsius = float( input('Ingresa una temperatura en °C'))
fahrenheit = (celsius * 9/5) + 32
print (f"{celsius}°C equivalen a {fahrenheit}°f")
```

Ejercicio 4. Escribe un programa que pida números al usuario y los vaya sumando. El ciclo termina cuando el usuario ingresa un `0` o cuando la suma es mayor que 50. Usa `while` y `break`.


```python
# Respuesta
suma_total = 0
while suma_total <= 50:
    numero = int(input(f"Suma actual [{suma_total}] - Ingresa un número: "))
    if numero == 0:
        print("BREAK] Ingresaste 0. Deteniendo el programa...")
    suma_total += numero
if suma_total > 50:
    print(f"¡Meta alcanzada! La suma final fue de {suma_total}, superando el límite de 50.")
else:
    print(f"El programa se cerró con una suma final de {suma_total}.")
```

Ejercicio 5. Usa un ciclo for para guardar en una lista los números del 1 al 10, pero salta los números pares usando continue.



```python
# Respuesta
numeros_impares = []
for i in range (1,11):
    if i % 2 == 0:
        continue
    numeros_impares.append(i)

print (f"La lista de numeros impares es: {numeros_impares}")
```

Ejercicio 6. Escriba una función que tenga como argumento el lado de un cuadrado y regrese su área.

```python
# Respuesta
def area_cuadrado (lado):
    area = lado * lado
    return area
lado_cuadrado = float(input('Ingresa cuanto mide el lado del cuadrado'))
resultado_area = area_cuadrado (lado_cuadrado)
print (f"Un cuadrado con lado {lado_cuadrado} tiene un área de: {resultado_area}")
```

Ejercicio 7. Escribe una función *my_max(x)* que de como salida el valor máximo (más grande) de una lista *x*. No use la función de Python *max*.

Encuentra el máximo de la siguiente lista 

x = [10, 56, 258, 16, 24, 18, 265, 893, 52, 39, 82]

```python
# Respuesta
def my_n (x):
    if len (x) == 0:
        return None
    maximo_actual = x [0]
    for numero in x:
        if numero > maximo_actual:
            maximo_actual= numero
    return maximo_actual
x = [10, 56, 258, 16, 24, 18, 265, 893, 52, 39, 82]
valor_maximo = my_n (x)
print (f"La lista analiza es {x}")
print (f"El valor maximo encontrado es: {valor_maximo}")
```

Ejercicio 8. Pide al usuario una contraseña. Permite hasta 3 intentos, y termina el ciclo si la acierta. Usa while, break y un contador.

```python
Password_correcto = "Python2026"
intentos = 0
max_intentos = 3
while intentos < max_intentos:
    password_usuario = input('Ingresa la contraseña:')
    intentos += 1
    if password_usuario == Password_correcto:
        print ('Acceso concedido')
        break

    else:
        intentos_restantes = max_intentos - intentos
        if intentos_restantes > 0:
            print (f"Contraseña incorrecta. Te quedan {intentos_restantes} intentos")
        else:
            print ("Acceso bloqueado")
```
