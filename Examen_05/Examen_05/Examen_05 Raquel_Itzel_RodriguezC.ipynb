**1.** Escribe una función llamada calcular_precio() que reciba:

El precio de un producto y un porcentaje de descuento.

La función debe regresar el precio final después del descuento.

Prueba la función con un producto de $500 sin indicar el descuento.

```python
# Respuesta
def calcular_precio(precio, descuento_porcentaje=0.10):
    precio_final = precio - (precio * descuento_porcentaje)
    return precio_final

producto = 500
precio_con_descuento = calcular_precio(producto)
print("El total de tu compra es:", precio_con_descuento)
```
