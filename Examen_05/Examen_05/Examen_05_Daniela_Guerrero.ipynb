**1.** Escribe una función llamada calcular_precio() que reciba:

El precio de un producto y un porcentaje de descuento.

Si no se especifica el descuento, debe asumir un valor de 10%.

La función debe regresar el precio final después del descuento.

Prueba la función con un producto de $500 sin indicar el descuento.

```python
# Respuesta
def calcular_precio(preciobase, descuento=10):
    Total=preciobase-(preciobase*(descuento/100))
    return Total
```

```python
preciobase=float(input('Introduce el precio base de tu producto: '))
Tot=calcular_precio(preciobase)
print('Tienes que pagar: ', Tot)
```
