**1.** Escribe una función llamada calcular_precio() que reciba:

El precio de un producto y un porcentaje de descuento.

Si no se especifica el descuento, debe asumir un valor de 10%.

La función debe regresar el precio final después del descuento.

Prueba la función con un producto de $500 sin indicar el descuento.

```python
# Respuesta
def calcular_precio(precio, descuento = 10):
  precio_final = precio * (1 - (descuento / 100))
  return precio_final

precio_og = float(input('Inserte el precio del pronducto: '))
descuento_conocido = input('¿Conoce usted el porcentaje del descuento? (Si/No)')
descuento_conocido.lower()
if descuento_conocido == 'si':
  descuento = float(input('Inserte el porcentaje de descuento: '))
  while descuento <= 1:
    descuento = float(input('Por favor, ingrese el número del descuento en porcentaje :'))
  precio_now = calcular_precio(precio_og, descuento)
  print(f'Su producto cuesta {precio_og} pesos y cuenta con un descuento de {descuento} %, \npor lo que el precio actual es {precio_now}')
else:
  precio_now = calcular_precio(precio_og)
  print(f'Su producto cuesta {precio_og} pesos y cuenta con un descuento de 10%, \npor lo que el precio actual es {precio_now}')




```

**2.**  Escribe una función llamada promedio() que pueda recibir cualquier cantidad de calificaciones usando *args.

La función debe regresar el promedio de todas las calificaciones recibidas.

Prueba la función con las calificaciones:

80, 90, 75, 95

e imprime el resultado.


```python
# Respuesta
def promedio(*args):
  value = 0
  for n in args:
    value += n
    prom = value / n
    return prom

prom = promedio(80, 90, 75, 95)
print(f'Entonces el promedio es {prom}')
```
