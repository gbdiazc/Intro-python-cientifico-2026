**1.** Escribe una función llamada calcular_precio() que reciba:

El precio de un producto y un porcentaje de descuento.

Si no se especifica el descuento, debe asumir un valor de 10%.

La función debe regresar el precio final después del descuento.

Prueba la función con un producto de $500 sin indicar el descuento.

```python
# Respuesta
def calcular_precio(precio,descuento=0.1):
    total=precio-(precio*descuento)
    return total
producto=500
costo=calcular_precio(producto)
print(costo)
```

**2.**  Escribe una función llamada promedio() que pueda recibir cualquier cantidad de calificaciones usando *args.

La función debe regresar el promedio de todas las calificaciones recibidas.

Prueba la función con las calificaciones:

80, 90, 75, 95

e imprime el resultado.


```python
# Respuesta
def promedio(*args):
    value=0
    suma=0
    for n in args:
        suma+=n
        value= suma/len(args)
    return value  
promedio(80, 90, 75, 95)
```
